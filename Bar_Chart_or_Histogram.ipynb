# **Import Library**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# **Create Sample Data**

```python
# For a categorical variable (e.g., gender)
data = {'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Female', 'Male', 'Female', 'Male']}
df = pd.DataFrame(data)

# For a continuous variable (e.g., age)
data_continuous = {'Age': [23, 45, 31, 22, 45, 56, 29, 30, 41, 55]}
df_continuous = pd.DataFrame(data_continuous)
```

# **Bar Chart**

```python
# Bar chart for categorical variable
plt.figure(figsize=(8, 6))
sns.countplot(x='Gender', data=df)
plt.title('Distribution of Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

# **Histogram**

```python
# Histogram for continuous variable
plt.figure(figsize=(8, 6))
sns.histplot(df_continuous['Age'], bins=5, kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```
